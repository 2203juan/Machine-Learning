```python
#borramos los atributos cuya correlacion es mayor a 0.8
def correlation(dataset, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
                if colname in dataset.columns:
                    del dataset[colname] # deleting the column from the dataset
                    print("borre "+colname)

    return dataset
data = correlation(data,0.8)
sns.heatmap(data.corr(), square=True, annot=True)
```

# Entrenamiento de modelos

```python
#Importar librerías necesarias para los dos modelos y las métricas de evluación

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import svm
import warnings
from sklearn.svm import SVC 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
import time
from sklearn import preprocessing
warnings.filterwarnings("ignore")
import statistics
from imblearn.over_sampling import RandomOverSampler
```

```python
# Realizamos el holdout


HOLDOUT = 10

h_precision = []
h_recall = []
h_f1score = []
h_accuracy = []

h_precision_svm = []
h_recall_svm = []
h_f1score_svm = []
h_accuracy_svm = []

start_time = time.time()

for i in range(HOLDOUT):

    # Separacion de los datos en conjuntos de entrenamiento y prueba
    X_train, X_test, y_train, y_test = train_test_split(data.drop(['F11'],axis=1), data['F11'], test_size=0.4)

    # Estimación de parámetros
    # En este código se ilustra el uso de la función que trae la librería sklearn para estimar paramentros,
    # la cual no necesita la extracción explícita del conjunto de validación pues hace la estimación usando
    # validación cruzada sobre el conjunto de entrenamiento.


    #----------------Se crea el clasificador MLP-------------------
    mlp = MLPClassifier()

    # Se definen los valores a explorar para cada parametro a estimar
    parameter_space = {
        'hidden_layer_sizes': [(i,) for i in range(1,21)],
        'activation': ['logistic', 'tanh', 'relu'],
    }

    # Se realiza la estimacion de parametros, en clf queda el modelo construido con los mejores
    # parametros encontrados y reentrenado con el conjunto de datos completo (refit)
    # Se hace la prediccion sobre los datos de prueba
    clf = GridSearchCV(mlp, parameter_space, n_jobs=-1, cv=3, refit=True)
    clf.fit(X_train,y_train)
    pred = clf.predict(X_test)

    # Se calculan metricas a partir de los datos de prueba PARA MLP
    mat = confusion_matrix(y_test, pred)
    print("Matriz de confusión MLP")
    print(mat)
    
    h_precision.append(precision_score(y_test,pred,average = "macro",labels=np.unique(pred)))
    h_recall.append(recall_score(y_test,pred,average = "macro",labels=np.unique(pred)))
    h_f1score.append(f1_score(y_test,pred,average = "macro",labels=np.unique(pred)))
    h_accuracy.append( accuracy_score(y_test,pred))
    
    #------Clasificador por maquinas de vectores de soporte------------------
    clf_svm = svm.SVC(kernel='linear')
    clf_svm.fit(X_train, y_train)
    pred_svm = clf_svm.predict(X_test)
    
    # Se calculan metricas PARA SVM
    mat_svm = confusion_matrix(y_test, pred_svm)
    print("Matriz de confusión SVM")
    print(mat_svm)
    
    h_precision_svm.append(precision_score(y_test,pred_svm,average = "macro",labels=np.unique(pred_svm)))
    h_recall_svm.append(recall_score(y_test,pred_svm,average = "macro",labels=np.unique(pred)))
    h_f1score_svm.append(f1_score(y_test,pred_svm,average = "macro",labels=np.unique(pred)))
    h_accuracy_svm.append(accuracy_score(y_test,pred_svm))

def fround(n):
    return str(n)[:10]

print("------------------------Reporte del Holdout------------------------")
print("-------------------------------MLP---------------------------------")
print("|-----Nombre-----|---------Media---------|-------Desviación-------|")
print("  Accuracy :            ",fround(statistics.mean(h_accuracy)),"            ",fround(statistics.stdev(h_accuracy)))
print("  Precision:            ",fround(statistics.mean(h_precision)),"            ",fround(statistics.stdev(h_precision)))
print("  Recall:               ",fround(statistics.mean(h_recall)),"            ",fround(statistics.stdev(h_recall)))
print("  F1score:              ",fround(statistics.mean(h_f1score)),"            ",fround(statistics.stdev(h_f1score)))
print("-------------------------------------------------------------------")

print("-------------------------------SVM---------------------------------")
print("|-----Nombre-----|---------Media---------|-------Desviación-------|")
print("  Accuracy :            ",fround(statistics.mean(h_accuracy_svm)),"            ",fround(statistics.stdev(h_accuracy_svm)))
print("  Precision:            ",fround(statistics.mean(h_precision_svm)),"            ",fround(statistics.stdev(h_precision_svm)))
print("  Recall:               ",fround(statistics.mean(h_recall_svm)),"            ",fround(statistics.stdev(h_recall_svm)))
print("  F1score:              ",fround(statistics.mean(h_f1score_svm)),"            ",fround(statistics.stdev(h_f1score_svm)))
print("-------------------------------------------------------------------")
tf = (time.time() - start_time)
print("Tiempo:  %s segundos " % tf)
print("Que son aproximadamente %s minutos" %(tf/60) )
print("Mejores parametros:")
print(clf.best_params_)
```

# Segunda Iteración

### Se hace oversampling para balancear el conjunto de datos

```python
data.F11.value_counts().plot(kind='bar', title='Count (F11) Estado Actual');
X = data.drop(['F11'],axis=1)
y = data["F11"]


ros = RandomOverSampler(random_state=0)
X_resampled, y_resampled = ros.fit_resample(X, y)

df = pd.DataFrame(X_resampled)
df["F11"] = y_resampled
```

```python
df.F11.value_counts().plot(kind='bar', title='Count (F11) Balanceados Con Oversampling');
```

### Realizamos la segunda iteración con el modelo seleccionado pero ahora con los datos balanceados y se añade a la estimación el parametro de penalización  (alpha)


### Se deciden hacer estas mejoras porque:
    -Los modelos van a poder hacer mejores inferencias y aprender mucho más sobre los datos si estos no están sesgados por la superioridad o inferioridad en tamaño de las diferentes clases que hay
    
    -La estimación de nuevos parámetros anteriormente no considerados es importante porque estos pueden aportar un mejor desempeño al modelo

```python
HOLDOUT = 10

h_precision = []
h_recall = []
h_f1score = []
h_accuracy = []


start_time = time.time()

for i in range(HOLDOUT):

    # Separacion de los datos en conjuntos de entrenamiento y prueba
    X_train, X_test, y_train, y_test = train_test_split(df.drop(['F11'],axis=1), df['F11'], test_size=0.4)

    # Estimación de parámetros
    # En este código se ilustra el uso de la función que trae la librería sklearn para estimar paramentros,
    # la cual no necesita la extracción explícita del conjunto de validación pues hace la estimación usando
    # validación cruzada sobre el conjunto de entrenamiento.


    #----------------Se crea el clasificador MLP-------------------
    mlp = MLPClassifier()

    # Se definen los valores a explorar para cada parametro a estimar
    parameter_space = {
        'hidden_layer_sizes': [(h,) for h in range(1,21)],
        'activation': ['logistic', 'tanh', 'relu'],
        'alpha': [0.0001, 0.05],
    }

    # Se realiza la estimacion de parametros, en clf queda el modelo construido con los mejores
    # parametros encontrados y reentrenado con el conjunto de datos completo (refit)
    # Se hace la prediccion sobre los datos de prueba
    clf = GridSearchCV(mlp, parameter_space, n_jobs=-1, cv=3, refit=True)
    clf.fit(X_train,y_train)
    pred = clf.predict(X_test)

    # Se calculan metricas a partir de los datos de prueba PARA MLP
    mat = confusion_matrix(y_test, pred)
    print("Matriz de confusión MLP")
    print(mat)
    
    h_precision.append(precision_score(y_test,pred,average = "macro",labels=np.unique(pred)))
    h_recall.append(recall_score(y_test,pred,average = "macro",labels=np.unique(pred)))
    h_f1score.append(f1_score(y_test,pred,average = "macro",labels=np.unique(pred)))
    h_accuracy.append( accuracy_score(y_test,pred))

def fround(n):
    return str(n)[:10]

print("------------------------Reporte del Holdout------------------------")
print("-------------------------------MLP---------------------------------")
print("|-----Nombre-----|---------Media---------|-------Desviación-------|")
print("  Accuracy :            ",fround(statistics.mean(h_accuracy)),"            ",fround(statistics.stdev(h_accuracy)))
print("  Precision:            ",fround(statistics.mean(h_precision)),"            ",fround(statistics.stdev(h_precision)))
print("  Recall:               ",fround(statistics.mean(h_recall)),"            ",fround(statistics.stdev(h_recall)))
print("  F1score:              ",fround(statistics.mean(h_f1score)),"            ",fround(statistics.stdev(h_f1score)))
print("-------------------------------------------------------------------")

tf = (time.time() - start_time)
print("Tiempo:  %s segundos " % tf)
print("Que son aproximadamente %s minutos" %(tf/60) )
print("Mejores parametros:")
print(clf.best_params_)
```

## Mejoras logradas y análisis de los resultados obtenidos

### Con el balanceo de los datos y la estimación del nuevo parámetro alpha se logra una mejora del 21% (pasamos de un desempeño de 65% a 86%)

Es evidente entonces que el balanceo de los datos es un factor importantísimo a la hora de entrenar nuestros modelos, pues de esta manera los modelos no estarán sesgados por las diferencias de tamaño entre las clases que existen en el conjunto de datos.Además se estimó que el mejor valor de alpha es 0.0001, el cual también contribuye a la mejoría del modelo en esta iteración, pues si bien es el valor que la librería pone por defecto, ya tenemos información valiosa: no debemos modificar el valor de alpha porque está en su valor óptimo

# Tercera Iteración

### Realizamos la tercera iteración sin estimacion de parametros nuevos (se usan los mejores de la iteracion anteior) y como  ultima mejora se decide aumentar la cantidad de unidades por capa oculta

    -Esta mejora se decide hacer gracias a la metodología de ensayo y error, pues en verdad eran pocas las modificaciones que se podían hacer.Una posibilidad que se tenía era hacer oversampling para aumentar el tamaño de todo el conjunto de datos y así entrenar mejor el modelo, pero como en la segunda iteración lo utilizamos para balancear los datos, decidí no hacerlo para esta.Es así como se llegó a la idea de aumentar la cantidad de unidades por capa oculta experimentalmente, obteniendo una mejora notoria en el desempeño del modelo.

```python
HOLDOUT = 10

h_precision = []
h_recall = []
h_f1score = []
h_accuracy = []


start_time = time.time()

for i in range(HOLDOUT):

    # Separacion de los datos en conjuntos de entrenamiento y prueba
    X_train, X_test, y_train, y_test = train_test_split(df.drop(['F11'],axis=1), df['F11'], test_size=0.4)

    # Estimación de parámetros
    # En este código se ilustra el uso de la función que trae la librería sklearn para estimar paramentros,
    # la cual no necesita la extracción explícita del conjunto de validación pues hace la estimación usando
    # validación cruzada sobre el conjunto de entrenamiento.


    #----------------Se crea el clasificador MLP-------------------
    clf = MLPClassifier(hidden_layer_sizes = (70,),activation = 'tanh',alpha = 0.0001)

    # Se hace la prediccion sobre los datos de prueba
    clf.fit(X_train,y_train)
    pred = clf.predict(X_test)

    # Se calculan metricas a partir de los datos de prueba PARA MLP
    mat = confusion_matrix(y_test, pred)
    print("Matriz de confusión MLP")
    print(mat)
    
    h_precision.append(precision_score(y_test,pred,average = "macro",labels=np.unique(pred)))
    h_recall.append(recall_score(y_test,pred,average = "macro",labels=np.unique(pred)))
    h_f1score.append(f1_score(y_test,pred,average = "macro",labels=np.unique(pred)))
    h_accuracy.append( accuracy_score(y_test,pred))

def fround(n):
    return str(n)[:10]

print("------------------------Reporte del Holdout------------------------")
print("-------------------------------MLP---------------------------------")
print("|-----Nombre-----|---------Media---------|-------Desviación-------|")
print("  Accuracy :            ",fround(statistics.mean(h_accuracy)),"            ",fround(statistics.stdev(h_accuracy)))
print("  Precision:            ",fround(statistics.mean(h_precision)),"            ",fround(statistics.stdev(h_precision)))
print("  Recall:               ",fround(statistics.mean(h_recall)),"            ",fround(statistics.stdev(h_recall)))
print("  F1score:              ",fround(statistics.mean(h_f1score)),"            ",fround(statistics.stdev(h_f1score)))
print("-------------------------------------------------------------------")

tf = (time.time() - start_time)
print("Tiempo:  %s segundos " % tf)
print("Que son aproximadamente %s minutos" %(tf/60) )
```
